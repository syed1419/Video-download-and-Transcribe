```python
from pytube import YouTube

def download_youtube_video(video_url, output_path='.'):
    try:
        yt = YouTube(video_url)
        video_stream = yt.streams.get_highest_resolution()
        video_title = yt.title
        if '/' in video_title:
          video_title = video_title.replace('/', '')
        video_title = f"{video_title}.mp4"
        # output_path = f"{output_path}/{video_title}.mp4"
        # print(output_path)

        print(f"Downloading {video_title}...")
        video_stream.download(output_path=output_path, filename=video_title)
        print(f"{video_title} has been downloaded successfully to {output_path}")

    except Exception as e:
        print(f"An error occurred: {e}")

    return video_title

if __name__ == "__main__":
    youtube_url = 'https://www.youtube.com/watch?v=mXGcBvWYl-U&ab_channel=edureka%21'

    output_directory = '/content/'

    video_title = download_youtube_video(youtube_url, output_directory)
```

```python
import moviepy.editor as mp

def convert_mp4_to_mp3(video_file, output_file):
    print(video_file)
    video = mp.VideoFileClip(video_file)
    audio = video.audio
    audio.write_audiofile(output_file)
    return output_file

if __name__ == "__main__":
    video_file = video_title
    audio_title = video_title.replace(".mp4", "")
    print(type(audio_title))
    output_file = f'{audio_title}.mp3'
    print(output_file)

    convert_mp4_to_mp3(video_file, output_file)
```
